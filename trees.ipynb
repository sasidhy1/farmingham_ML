```python
import pandas as pd
```

```python
df = pd.read_csv('reduced.csv')
df.head()
```

```python
target = df[["ANGINA"]]
# target_names = ["negative", "positive"]
```

```python
data = df.drop(["ANGINA"], axis=1)
feature_names = data.columns
data.head()
```

```python
print(data.shape, target.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

### decision tree

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
sorted(zip(clf.feature_importances_, feature_names), reverse=True)[:10]
```

```python
predictions = clf.predict(X_test)
```

```python
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test.values.ravel()}).reset_index(drop=True)
results.replace({0:'negative',1:'positive'}).head(10)
```

### random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train.values.ravel())
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)[:10]
```

```python
predictions = rf.predict(X_test)
```

```python
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test.values.ravel()}).reset_index(drop=True)
results.replace({0:'negative',1:'positive'}).head(10)
```
